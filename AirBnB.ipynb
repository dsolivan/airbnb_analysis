##Cleaning Data

###Cleaning Empty Cells
We want to drop rows with null values but ONLY for values in price column since other columns are less important if they have null values.

```python
#Drop null values in price column
df.dropna(subset=['price'],inplace=True)
#Check if null values have been removed
df.isnull().sum()
```

###Cleaning Unnecessary Columns

```python
#Drop id and name
df.drop(columns=['id','name'], inplace=True)
df.info()
```

###Removing Duplicates

```python
#Drop Duplicates by host_id, neighbourhood_group, neighbourhood, and price
df.drop_duplicates(subset=['host_id','neighbourhood','neighbourhood_group','price'], keep='first', inplace=True)
```

###Removing Outliers

```python
#Checking outliers on price
plt.boxplot(df['price'], vert=False)
plt.title("Boxplot for Outlier Detection")
plt.xlabel("Price")
plt.show()
```

```python
df_sorted = df.sort_values(by='price',ascending=False)
df_sorted.head(10)
```

We see there are a few outliers in our data. Lets remove rows with a price >= 10000

```python
#Keeps rows with a price less than $10,000
df = df[df['price'] < 10000]
```

##Exploring Data

```python
#Creating a Correlation Matrix and Numerizing Object Columns
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
correlation_matrix = df_numerized.corr()

#Heatmap Matrix
sns.heatmap(correlation_matrix, annot = True)

plt.show()
```
